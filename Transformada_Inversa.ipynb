**Distribución Triangular (Método de la Transformada Inversa)**


La distribución triangular fue una de las primeras distribuciones propuestas de tipo continuo. Se caracteriza por tener una aproximación inicial en aquellas situaciones donde se carece de información.

Su función de densidad es:

$$
f(x) = \begin{cases}
\frac{2}{(c-a)(b-a)} (x-a) & \text{si } \text{ } a \leq x \leq b \\
\frac{-2}{(c-a)(c-b)}(x-c) & \text{si } \text{ }  b < x \leq c
\end{cases}
$$


Su función de distribución es:

$$ F(x)=
\begin{cases}
    \frac{(x-a)^2}{(c-a)(b-a)} & \text{si } a \le x \le b \\
    1 + \frac{(c-x)^2}{(c-a)(c-b)} & \text{si } b < x \le c
\end{cases}
$$



La inversa de la función de distribución es:  

$$  F^{-1}(R)=
\begin{cases}
    a + \sqrt{R(b-a)(c-a)} , \text{ si } \text{ }  R < \frac{b-a}{c-a}\\
    c - \sqrt{(1-R)(c-b)(c-a)}, \text{ en otro caso}
\end{cases}
$$

```python
#Importamos las librerias
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
#Proporcionamos los datos de a,b y c
a = 1
b = 3
c = 5
```

```python
#Definimos la función de distribución inversa

def distribucion_inversa(n):
  lista = [] #Creamos una lista
  for i in range(n): #Generamos un ciclo for
      R = rd.random() # #Generamos un número aleatorio
    #Evaluamos los casos de la distribución
    #Si R < (b-a)/(c-a)
      if R < (b-a)/(c-a):
          x = a + np.sqrt(R*(c-a)*(b-a))
      else:
    #En otro caso
        x = c - np.sqrt((1-R)*(c-b)*(c-a))
      lista.append(x)
  return lista
```

```python
#Realizamos el experimento
l = distribucion_inversa(10000)
```

```python
print(l)
```

```python
plt.hist(l, bins=50, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Distribución Triangular")
plt.xlabel("x")
plt.ylabel("Densidad")
plt.grid(True, linestyle ='--', alpha = 0.5)
plt.show()
```

```python
print("La media es :", np.mean(l))
print("La varianza es: ", np.var(l))
print("La desviación estándar es: ", np.std(l))

print("Los resultados teóricos:")
print("La media es: ",(a+b+c)/3)

print("La varianza es: ",((a**2 + b**2 + c**2 - a*c - a*b - b*c)/18))

print("La desviación estandar es: ", np.sqrt((a**2 + b**2 + c**2 - a*c - a*b - b*c)/18))
```
