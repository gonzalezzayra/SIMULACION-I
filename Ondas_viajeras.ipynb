Grafique la función $f(x')=\exp\left( -\frac{(x'-0.5)^2}{0.01} \right)$ y suponga que esta representa una pertubación en una cuerda.

a). Mostrar que al hacer al cambio de variable $x' = x-ct$ la perturbación se desplaza a la derecha.

b). Mostrar que al hacer el cambio de variable $x' = x+ct$ la perturbación se desplaza a la izquierda.

c). Graficar la superposición de las ondas viajeras de los incisos a) y b)

```python
#importamos librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definimos la función f(x') = exp(-((x' - 0.5)^2) / 0.01)
def f(xp):
    return np.exp(-((xp - 0.5)**2) / 0.01)
```

```python
c = 1 #valor arbitrario
#Definimos la superposición de las ondas viajeras
#Coordenadas de traslación
#x' = x + ct, izquierda
#x' = x - ct derecha

#De lado derecho x' = x -ct
def f_derecha(x, t, c):
    return f(x - c*t)

#De lado izquierdo x' = x + ct
def f_izquierda(x, t, c):
    return f(x + c*t)

#La superposición
def f_total(x, t, c):
    # x' = x - ct (derecha) y z = x + ct (izquierda)
    return f(x - c*t) + f(x + c*t)
```

```python
#Para graficar
x = np.linspace(0, 1, 1000)  #1000 valores entre 0 y 1
tiempos = [0.3]  # Tiempos t = 0.3

#Graficamos
plt.figure(figsize=(10, 6))
for t in tiempos:
    y = f_derecha(x, t, c)
    plt.plot(x, y, label=f't = {t}')

# Configuramos la gráfica
plt.title('Superposición de ondas viajeras (lado derecho)')
plt.xlabel('x')
plt.ylabel('f(x, t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Para graficar
x = np.linspace(0, 1, 1000)  #1000 valores entre 0 y 1
tiempos = [0.3]  # Tiempos t = 0.3

#Graficamos
plt.figure(figsize=(10, 6))
for t in tiempos:
    y = f_izquierda(x, t, c)
    plt.plot(x, y, label=f't = {t}')

#Configuramos la gráfica
plt.title('Superposición de ondas viajeras (lado izquierdo)')
plt.xlabel('x')
plt.ylabel('f(x, t)')
plt.legend()
plt.grid(True)
plt.show()
```
