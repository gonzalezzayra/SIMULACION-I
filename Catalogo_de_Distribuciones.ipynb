**Uniforme**


----

**Nomenclatura**

$$X \sim U(a,b)$$

**Expresiones téoricas para la media, la varianza y desviación estándar**

\begin{array}{|c|c|}
\hline
\text{Rango} & [a,b] \\
\hline
\text{Media} & \frac{a+b}{c} \\
\hline
\text{Varianza} & \frac{(b-a)^2}{12}  \\
\hline
\end{array}




**PDF:**

$$
f(x) =
\begin{cases}
    \frac{1}{b-a} & \text{si } a \leq x \leq b  \\
    0   & \text{en otro caso }
\end{cases}
$$

**CDF:**

$$
F(x) =
\begin{cases}
    0 & \text{si } x < a \\
    \frac{x-a}{b-a}   & \text{si } a \leq  x  \leq  \\
    1 & \text{si } b < x
\end{cases}
$$


IMG_0636.png

**Pseudocódigo**

\begin{aligned}
&\text{Generar } U \sim U(0,1) \\
&\text{Hacer } X = a + (b-a)\cdot U\\
&\text{Salir } X
\end{aligned}

**Código**

```python
#Uniforme
import numpy as np
import matplotlib.pyplot as plt

def simular_uniforme(a, b, n=1000):
    return np.random.uniform(a, b, n)

#Parámetros y simulación
a=0
b=1
muestras = simular_uniforme(a, b)

#Valores
print('La varianza es:',np.var(muestras))
print('La media es:',np.mean(muestras))
print('La varianza teorica es:', (b-a)**2/12)
print('La media teorica es:', (a+b)/2)


#Configurar gráficos
plt.figure(figsize=(12, 4))

#Histograma (PDF)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='skyblue', edgecolor='black')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('PDF de Uniforme({}, {})'.format(a, b))
```

**Triangular**

---


**Nomenclatura**

$$X \sim Tr(a,b,c)$$

\begin{array}{|c|c|}
\hline
\text{Rango} & [a,c] \\
\hline
\text{Media} & \frac{a+b+c}{3} \\
\hline
\text{Varianza} & \frac{a^2 + b^2 + c^2 - ab - ac -bc }{18}  \\
\hline
\end{array}

**PDF:**
$$
f(x) =
\begin{cases}
    \frac{2(x-a)}{(c-a)(b-a)} & \text{si } a \leq x \leq b \\
    \frac{-2(x-c)}{(c-a)(c-b)}   & \text{si } b < x \leq c  
\end{cases}
$$

**CDF:**

$$
F(x) =
\begin{cases}
    \frac{(x-a)^2}{(c-a)(b-a)} & \text{si }  a \leq x \leq b  \\
     1 + \frac{(x-c)^2}{(c-a)(b-c)}  & \text{si } b < x \leq c
\end{cases}
$$


IMG_0637.png

**Pseudocódigo**

\begin{aligned}
&\text{Generar } U \sim U(0,1) \\
&\text{Hacer si  } U \leq \frac{b-a}{c-a}: \\
&\text{ } X= a + \sqrt{ U (b-a) (c-a)} \\
&\text{En otro caso: }  \\
&\text{ }  X = c - \sqrt{(1-U)(c-b)(c-a)}\\
&\text{Salir } X
\end{aligned}

**Código**

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np

a = 1
b = 3
c = 5

def tri(n):
  x = []
  for i in range(n):
    R = rd.random()
    if  R <(b-a)/(c-a):
      x.append(a + np.sqrt(R * (b - a) * (c - a)))
    else:
      x.append(c - np.sqrt((1 - R) * (c - b) * (c - a)))
  return x

k = tri(100000)

plt.hist(k, bins=100, color = 'skyblue') #Histograma de una distribución triangular
plt.title("Histograma de una Distribución Triangular")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

**Exponencial**

---

**Nomenclatura**

$$X\sim \exp(\beta)$$

**Expresiones téoricas para la media, la varianza y desviación estándar**

\begin{array}{|c|c|}
\hline
\text{Rango} & [0,∞) \\
\hline
\text{Media} & β \\
\hline
\text{Varianza} & β^2\\
\hline
\end{array}

**PDF & CDF**

**PDF:**

$$
f(x) =
\begin{cases}
    \frac{1}{\beta}e^{-x/\beta} & \text{si } x \geq 0 \\
    0   & \text{en otro caso }  
\end{cases}
$$


**CDF:**

$$
F(x) =
\begin{cases}
    1 - e^{-x/\beta} & \text{si } x \geq 0 \\
    0   & \text{en otro caso }  
\end{cases}
$$


IMG_0638.png

**Pseudocódigo:**

\begin{aligned}
&\text{Generar } U\sim U(0,1) \\
&\text{Hacer } X = -\beta \cdot\ln U \\
&\text{Salir } X
\end{aligned}

**Código**

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np

beta = 1

def exp(n):
    lista = []
    for i in range(n):
        R = rd.random()
        x = - beta*np.log(R)
        lista.append(x)
    return lista

l = exp(100000)

plt.hist(l, bins=100, color = 'skyblue') #Histograma de una distribución triangular
plt.title("Histograma de una Distribución Exponencial")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

**m-Erlang**

---

**Nomenclatura**

$$X \sim \text{Erlang} (m,λ)$$

**Expresiones téoricas para la media, la varianza y desviación estándar**

\begin{array}{|c|c|}
\hline
\text{Parámetros} & m \in Z^{+}, λ > 0 \\
\hline
\text{Media} & \frac{m}{λ} \\
\hline
\text{Varianza} & \frac{m}{λ^2}  \\
\hline
\end{array}

IMG_0635.png

**Pseudocódigo:**

\begin{aligned}
&\text{Generar } U_1, U_2, ..., U_m  \text{ como iid } U(0,1) \\
&\text{Hacer }
X = -\frac{\beta}{m} \ln\left( \prod_{i=1}^{m} U_i \right)
\\
&\text{Salir } X
\end{aligned}

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def simular_erlang(m, lamda, n=1000):
    U = np.random.rand(n, m)
    E = -np.log(U)/lamda
    X = np.sum(E, axis=1)
    return X

#Parámetros
m = 3
lamda = 0.5
muestras = simular_erlang(m, lamda)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='skyblue', edgecolor='black')
x = np.linspace(0, 20, 1000)
plt.plot(x, stats.erlang.pdf(x, m, scale=1/lamda), 'r-', lw=2)
plt.title(f'PDF m-Erlang(m={m}, λ={lamda})')
plt.xlabel('Valor')
plt.ylabel('Densidad')
```

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np

m = 3
labda = 1

def erlang(n):
    lista = []
    for i in range(n):
        x = 0
        for i in range(m):
            R = rd.random()
            x += -np.log(R) / labda
        lista.append(x)
    return lista

l = erlang(10000)

plt.hist(l, bins=100) #Histograma de una distribución erlang
plt.title("Histograma de una Distribución Erlang")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

**Gamma**

----

**Nomenclatura**

$$X\sim\text{gamma}(α,β)$$

**Expresiones téoricas para la media, la varianza y desviación estándar**

\begin{array}{|c|c|}
\hline
\text{Rango} & [0,∞] \\
\hline
\text{Media} & αβ \\
\hline
\text{Varianza} & αβ^{2}  \\
\hline
\end{array}

**PDF:**

$$
f(x) =
\begin{cases}
     \frac{\beta^{-\alpha} x^{\alpha -1} e^{-x/\beta}}{𝚪(\alpha)} & \text{si } x >  0 \\
    0   & \text{en otro caso }
\end{cases}
$$


**CDF:**

$$
F(x) =
\begin{cases}
    1 - e^{-x/\beta} \sum_{j=0}^{\alpha-1} \frac{(x/\beta)^{j}}{j!} & \text{si } x > 0 \\
    0   & \text{en otro caso }
\end{cases}
$$

IMG_0635.png

**Pseudocódigo:**

\begin{aligned}
&\text{Establecer } \alpha \text{ y } \beta \\
&\text{ X = 0 }\\
&\text{Para i desde 1 hasta } \alpha \\
&\text{Generar } U \sim U(0,1) \\
&\text{Hacer } E = - \ln (U)/\beta \\
&\text{Hacer }  X = X + E\\
&\text{Salir } X
\end{aligned}

**PDF:**

$$
f(x) =
\begin{cases}
    α β^{-α}x^{α-1} e^{(x/β)^{α}} & \text{si } x > 0 \\
    0   & \text{en otro caso }  
\end{cases}
$$


**CDF:**

$$
F(x) =
\begin{cases}
   1 - e^{-(x/β)^{α}}  & \text{si } x > 0 \\
    0   & \text{en otro caso }
\end{cases}
$$

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from math import gamma

def simular_gamma(alpha, beta, n=1000):
    muestras = np.zeros(n)
    for i in range(n):
        X = 0
        for _ in range(alpha):  # Suma α exponenciales
            U = np.random.rand()
            E = -np.log(U)/beta
            X += E
        muestras[i] = X
    return muestras

# Parámetros y simulación
alpha, beta = 3, 0.5  # Ejemplo: α entero
muestras = simular_gamma(alpha, beta, 10000)

# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PDF empírica)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='skyblue',
         edgecolor='black', alpha=0.7, label='Empírica')
```

**Weibull**

---


**Nomenclatura**

$$X\sim\text{Weibull}(α,β)$$

**Expresiones téoricas para la media, la varianza y desviación estándar**


\begin{array}{|c|c|}
\hline
\text{Rango} & [0, \infty) \\
\hline
\text{Media} & \dfrac{\beta}{\alpha} \cdot \Gamma\left( \dfrac{1}{\alpha} \right) \\
\hline
\text{Varianza} & \dfrac{\beta^2}{\alpha} \left\{ 2 \Gamma\left( \dfrac{2}{\alpha} \right) - \dfrac{1}{\alpha} \left[ \Gamma\left( \dfrac{1}{\alpha} \right) \right]^2 \right\} \\
\hline
\end{array}

**PDF:**

$$
f(x) =
\begin{cases}
    α β^{-α}x^{α-1} e^{(x/β)^{α}} & \text{si } x > 0 \\
    0   & \text{en otro caso }  
\end{cases}
$$


**CDF:**

$$
F(x) =
\begin{cases}
   1 - e^{-(x/β)^{α}}  & \text{si } x > 0 \\
    0   & \text{en otro caso }
\end{cases}
$$

IMG_0640.png

**Pseudocódigo:**

\begin{aligned}
&\text{Establecer } α \text{ y } \beta \\
&\text{Generar } U \sim U(0,1) \\
&\text{Hacer } X = \beta (-\ln U)^{1/\alpha}\\
&\text{Salir } X
\end{aligned}

**Código**

```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np

beta = 1.5
alphas = 2

def weibull(n):
    lista = []
    for i in range(n):
        R = rd.random()
        x = beta * (-np.log(R))**(1 / alphas)
        lista.append(x)
    return lista

c = weibull(1000000)


plt.hist(c, bins=2000) #Histograma de una distribución weibull
plt.title("Histograma de una Distribución Weibull")
plt.xlabel("")
plt.ylabel("Densidad de probabilidad")
plt.grid(True, linestyle='--', alpha=0.5)
```

**Normal**

----

**Nomenclatura**

$$X \sim (\mu,\sigma^2)$$

**Expresiones téoricas para la media, la varianza y desviación estándar**

\begin{array}{|c|c|}
\hline
\text{Rango} & (-∞, ∞) \\
\hline
\text{Media} & μ \\
\hline
\text{Varianza} & \sigma^2  \\
\hline
\end{array}

**PDF:**

$$
f(x) =
\frac{1}{\sqrt{2\pi\sigma^{2}}} e^{-\frac{(x-\mu)^2}{(2\sigma^2)}}
$$




IMG_0641.png

**Pseudocódigo:**

\begin{aligned}
&\text{Generar } U_1, U_2 \sim U(0,1) \\
&\text{Hacer } R = \sqrt{ -2\ln U_1 }, \quad \theta = 2\pi U_2 \\
&\text{Hacer } X = R \cos \theta, \quad Y = R \sin \theta \\
&\text{Salir } X, Y
\end{aligned}

**Código**

```python
#Importar librerias
import numpy as np
import matplotlib.pyplot as plt
import random as rd

#Definir box,como todos los calculos que se necesitan
def box():
    U1 = rd.uniform(0,1) #Generar los dos numeros aleatorios U1 y U2
    U2 = rd.uniform(0,1)
    R = np.sqrt(-2*np.log(U1)) #Calcular R
    theta = 2*np.pi*U2 #Calcular theta
    X = R*np.cos(theta) #Calcular X
    Y = R*np.sin(theta) #Calcular Y
    return X,Y

#Crear dos listas para x e y
listax = []
listay = []

#Definir el experimento
def experimento(n):
  for _ in range(n): #Ciclo for para que podamos generar varios
    X,Y = box() #Asignamos el valor de x e y a la función
    listax.append(X) #Adjuntar a la lista de x
    listay.append(Y) #Adjuntar a la lista de y
  return listax, listay

#experimento
l = experimento(100000)

plt.hist(l, bins=100 ,alpha = 0.7)
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.show()
```

**Log-normal**

---

**Nomenclatura**

$$X\sim LN(\mu,\sigma^2)$$

**Expresiones téoricas para la media, la varianza y desviación estándar**

\begin{array}{|c|c|}
\hline
\text{Rango} & [0,∞) \\
\hline
\text{Media} & e^{\mu + \sigma^{2}/2} \\
\hline
\text{Varianza} & e^{2\mu + \sigma^{2}} (e^{\sigma^2} -1)  \\
\hline
\end{array}

**PDF:**

$$
f(x) =
\begin{cases}
    \frac{1}{x \sqrt{2\pi\sigma^{2}}} e^{\frac{-(\ln x - \mu)^2}{2\sigma^2}} & \text{si } x > 0 \\
    0   & \text{en otro caso }  
\end{cases}
$$



IMG_0642.png

**Pseudocódigo:**

\begin{aligned}
&\text{Generar } Y \sim N(\mu,\sigma^{2}) \\
&\text{Hacer } X = e^{Y} \\
&\text{Salir } X
\end{aligned}

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, sqrt, exp
from scipy.special import erf  # Importar erf desde scipy.special

def box_muller(n=1000):
    m = (n + 1) // 2  # Para asegurar suficientes muestras

    U1 = np.random.rand(m)
    U2 = np.random.rand(m)

    Z0 = np.sqrt(-2 * np.log(U1)) * np.cos(2 * pi * U2)
    Z1 = np.sqrt(-2 * np.log(U1)) * np.sin(2 * pi * U2)

    return np.concatenate([Z0, Z1])[:n]

def log_normal(mu, sigma, n=1000):
    Z = box_muller(n)
    X = np.exp(mu + sigma * Z)
    return X

# Parámetros y simulación
mu = 0      # Parámetro mu de la log-normal
sigma = 1   # Parámetro sigma de la log-normal
muestras = log_normal(mu, sigma, 10000)  # 10,000 muestras log-normales


# Cálculos estadísticos

# Momentos teóricos de la log-normal
teorica_media = exp(mu + sigma**2/2)
teorica_varianza = (exp(sigma**2) - 1) * exp(2*mu + sigma**2)


# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PDF empírica)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='skyblue',
         edgecolor='black', alpha=0.7, label='Empírica', range=(0, 10))
```

**Beta**

----

**Nomenclatura**

$$X\sim beta(\alpha_1,\alpha_2)$$

**Expresiones téoricas para la media, la varianza y desviación estándar**

\begin{array}{|c|c|}
\hline
\text{Rango} & [0,1] \\
\hline
\text{Media} & \frac{\alpha_1}{\alpha_1 + \alpha_2} \\
\hline
\text{Varianza} & \frac{\alpha_1 \alpha_2}{ (\alpha_1 + \alpha_2)^2 (a_1 + a_2 +1 )}  \\
\hline
\end{array}

**PDF:**

$$
f(x) =
\begin{cases}
    \frac{x^{{\alpha_1} -1} (1-x)^{\alpha_2 -1} }{B(\alpha_1, \alpha_2)} & \text{si } 0 <  x < 1 \\
    0   & \text{en otro caso }  
\end{cases}
$$



IMG_0643.png

**Pseudocódigo:**

\begin{aligned}
&\text{Generar } Y_1 \sim gamma(\alpha, 1), Y_2 \sim gamma(\alpha_2,1) \text{ independiente de } Y_1 \\
&\text{Hacer } X = Y_1 / (Y_1 + Y_2) \\
&\text{Salir } X
\end{aligned}

**Código**

```python
import numpy as np
import matplotlib.pyplot as plt
from math import gamma
from scipy.stats import beta as beta_dist

def beta_pdf(x, alpha, beta):

    return (x**(alpha-1) * (1-x)**(beta-1)) / (gamma(alpha) * gamma(beta) / gamma(alpha + beta))

def simular_beta_ar(alpha, beta, n=1000):

    muestras = []
    M = beta_pdf((alpha-1)/(alpha+beta-2), alpha, beta)  # Máximo de f(x)

    while len(muestras) < n:
        X = np.random.uniform(0, 1)  # Distribución propuesta (Uniforme)
        U = np.random.uniform(0, 1)  # Uniforme para la condición

        if U <= beta_pdf(X, alpha, beta) / M:
            muestras.append(X)

    return np.array(muestras)

# Parámetros y simulación
alpha, beta = 2, 5
muestras = simular_beta_ar(alpha, beta, 10000)  # 10,000 muestras Beta(2,5)


# Gráficos

plt.figure(figsize=(12, 5))

# Histograma (PDF empírica)
plt.subplot(1, 2, 1)
plt.hist(muestras, bins=50, density=True, color='skyblue',
         edgecolor='black', alpha=0.7, label='Empírica', range=(0, 1))
```

$$
f(x) = \begin{cases}
\dfrac{\alpha_2(b - a)}{(x - a)(b - x)\sqrt{2\pi}} \exp\left( -\dfrac{1}{2} \left[ \alpha_1 + \alpha_2 \ln\left( \dfrac{x - a}{b - x} \right) \right]^2 \right), & \text{si } a < x < b \\
0, & \text{en otro caso}
\end{cases}
$$

**CDF:**

$$
F(x) = \begin{cases}
\Phi\left( \alpha_1 + \alpha_2 \ln\left( \dfrac{x - a}{b - x} \right) \right), & \text{si } a < x < b \\
0, & \text{en otro caso}
\end{cases}
$$

$$\text{donde } \Phi(x) \text{ es la función de distribución de una variable aleatoria normal con } \mu=0 \text{ y } \sigma^2 = 1$$

\begin{aligned}
&\text{Generar } Z \sim N(0,1) \\
&\text{Hacer } Y = exp[(Z - \alpha_1)/\alpha_2] \\
&\text{Hacer } X = (a + bY)/(Y+1) \\
&\text{Salir } X
\end{aligned}
