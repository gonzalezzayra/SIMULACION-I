**Método de los 12 uniformes**



**Para generar:**

$$f(x) = \frac{1}{\sqrt{2\pi} \sigma} e^{-\frac{1}{2}{}(\frac{x-\mu}{\sigma})^{2}}, -∞ < x < ∞ $$

**Usamos el teorema del límite central:**

Generamos n números aleatorios y consideramos

$$z = \frac{Y - \sum_{i=1}^{n} \mu_{i}}{\sqrt{\sum_{i=1}^{n} \sigma_{i}^{2}} }, \text{ } \text{ }  Y = R_1 + R_2 + ...+ R_n$$


$$
\therefore \quad
z = \frac{\sum_{i=1}^{n} R_i - \frac{n}{2}}{\sqrt{\frac{n}{12}}}
$$



con $n=12$

$$ z = \sum_{i=1}^{n} R_i -6 $$

$x = \mu + z\cdot \sigma$, sustituimos el valor de z en x

$$x = \mu + \sigma \cdot \sum_{i=1}^{n} R_i -6 $$

$$
x = \mu + \sigma ⋅\bigg( \sum_{i=1}^{12} R_i - 6 \bigg) \text{ } \text{ } R_i \sim U(0,1)
$$




En este ejercicio tenemos los parametros de

$$ \mu = 2 \text{ y } \sigma = 3  $$

```python
#Importar librerias
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generar 12 numeros aleatorios
n=12
#Parametros
mu = 2
desviacion = 3

#Definir la variable sumatoria que es z, la sumatoria de los 12 numeros aleatorios y restarle 6
def sumatoria():
  suma = 0
  for _ in range(n): #Ciclo for para generar los 12 numeros aleatorios y sumarlos
    numeros = rd.random()
    suma += numeros
  return suma - 6 #Restarle 6 a la suma de los numeros aleatorios

print("El valor de z es:", sumatoria()) #Imprimir el valor de z
```

```python
#Calcular x

lista= [] #Creamos una lista
def experimento(m): #Definimos la variable experimento con parametro m, para que no se confunda con n
  for _ in range(m): #Generamos ciclo for para calcular diferentes x's
    x = mu + desviacion*sumatoria()
    lista.append(x)
  return lista
```

```python
#Experimento con m parametro
l = experimento(100000)

#Graficar
plt.hist(l, bins=100, color = 'limegreen', edgecolor = 'black',alpha = 0.7)
plt.ylabel("Frecuencia")
plt.title("Método de las 12 uniformes")
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.show()
```

```python
#Media y desviación estandar del programa

print("La media es: ", np.mean(l))
print("La desviación estandar es: ", np.std(l))

#Resultados teóricos
print("Los resultados teóricos son: ")
print("La media es: ", mu)
print("La desviación estandar es: ", desviacion)
```
