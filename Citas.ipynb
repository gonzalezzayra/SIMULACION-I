Dos personas han quedado citas en un determinado lugar. Ambas tiene la misma probabilidad de llegar al lugar de la cita en el intervalo de tiempo [0,30] en minutos.
Determínese la probabilidad de que el tiempo que una persona espere a la otra sea como máximo 10 minutos.

```python
#Importar librerias
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definimos la función tiempos para los tiempos de cada persona
def tiempos(n):
    exitos = 0  #contador para el numero de exitos
    for i in range(n):
        #Generamos los numeros aleatorios entre 0 y 30. [0,30]
        t1 = random.uniform(0, 30) #Para la primera persona
        t2 = random.uniform(0, 30) #Para la segunda persona

        #Calcular el tiempo de espera
        #Como no puede haber tiempos negativos, se utiliza el valor absoluto
        tiempoesp = abs(t1 - t2)

        #Necesitamos ver si es un éxito
        if tiempoesp <= 10: #Si es menor que 10, se le suma un éxito
            exitos += 1
    prob = exitos / n #Obtenemos la probabilidad
    return prob
```

```python
tiempos(1000) #Calculamos para n=1000 como ejemplo
```

```python
#Definimos la función experimento
def experimento(m):
    lista = []  #Generamos la lista
    exitos_totales = 0  #Contador para los exitos totales
    #Con la estructura anterior de la función tiempos
    for i in range(m):  #Generamos Ciclo for con parametro m
        t1 = random.uniform(0, 30) #Para la primera persona
        t2 = random.uniform(0, 30) #Para la segunda persona
        tiempoesp = abs(t1 - t2) #Tiempo de espera
        if tiempoesp <= 10: #Vemos si es un exito o no
            exitos_totales += 1  #Los exitos totales
        prob_ac = exitos_totales/(i + 1) #Calcular probabilidad acumulada
        lista.append(prob_ac) #Se adjunta en la lista
    return lista #Se devuelve la lista
```

```python
#print(experimento(1000)) #solo era para ver la lista
```

```python
#Definimos el experimento con m = 1000
l = experimento(1000)
```

```python
iteraciones = list(range(1,1001))  #Lista de iteraciones para el eje x, se necesitaba definir esta lista
#Grafica
plt.plot(iteraciones, l, label='Probabilidad estimada')
plt.axhline(y=5/9, color='green', linestyle='--', label='Valor teórico: 5/9')
plt.xlabel('Número de simulaciones')
plt.ylabel('Probabilidad')
plt.title('Citas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Para obtener la probabilidad, tomamos el último valor de la lista
probabilidad = l[-1]
print("La probabilidad es :", probabilidad)
```
