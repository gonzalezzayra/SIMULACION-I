**Generador Congruencial**

$$x_{n+1}=(a x_n + c)\hspace{2pt} mod\hspace{2pt} m$$

En este código vamos a proporcionar un ejemplo.
Los valores de
$a=81$, $c=89$, $x_0 = 5$ y $m=10000$.



$x_0$ es la semilla

$m$ determina el ciclo de vida, a lo más hay m residuos




```python
import tabulate #tabulate imprime tablas a partir de los datos de entrada de su primer argumento


a = 81; # cambiar valores a y c
c = 89;
m = 100**2;
x0 = 5; #semilla
n = 10002; #damos un numero mayor a m ya que queremos ver hasta donde llega el ciclo de vida
#n es la cantidad de numeros que generamos
#Función generadora

#Definimos la función generadora junto con los parametros
def generador(a, c, m, x0):
  x = (a*x0 + c) % m #es la función del generador congruencial x_{n+1}=(a x_n +c) mod m
  return x # nos regresa el nuevo valor de la función

muestra =[] #generamos una lista para guardar los valores i, x_1,
lista=[] #generamos esta lista para guardar los valores normalizados

#Se utiliza un ciclo for para que nos genere los 'n' numeros
for i in range(n): #genera hasta 10002 veces por el dato proporcionado
  nuevo = generador(a,c,m,x0) #almacena el nuevo numero
  muestra.append([i,x0,nuevo/m]) #guarda en 'muestra' i, x_1 y el valor normalizado
  lista.append(nuevo/m) # guarda en 'lista' el valor normalizado
  x0=nuevo #se actualiza x_0 para el siguiente
encabezados=["i","x_1","x_{i+1}"] #se fijan los encabezados de la tabla
print(tabulate.tabulate(muestra,headers = encabezados)) # se imprime la tabla junto con los encabezados


```

En este ejemplo podemos observar que de la iteración 0 a la 9999 no se repite ningun termino, ya que el ciclo de vida es de $10000$.


$m$ determina el ciclo de vida, el valor de m fue de $100^2 = 10000$
