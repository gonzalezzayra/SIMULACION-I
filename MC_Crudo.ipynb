**MC Crudo(Media Muestral)**

```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Definimos la función de la raiz de arctan(x)
def f(x):
  return np.sqrt(np.arctan(x))
```

```python
#Generamos 10 números entre 0 y 1
x = np.linspace(0,1,10)
print(x)
```

```python
y = f(x)
print(y)
```

```python
#Definir la función de Montecarlo Crudo
def montecarlo(n):
  sumatoria = 0 #contador
  for i in range(n):
    x = rd.random() #se produce un numero aleatorio
    sumatoria += f(x)
  return sumatoria/n #es la sumatoria entre n
```

```python
#Tamaño de muestra
n = 10000

monte = montecarlo(n)

#Imprimir el tamaño de la muestra y el promedio/media
print("Con una muestra de", n , "se obtuvo", monte)
```

Definimos el experimento

```python
def experimento(m):
  lista = []
  for i in range(m):
    mont = montecarlo(i+1)
    lista.append(mont)
  return(lista)
```

```python
l = experimento(1000)
#print(l)

#se crea un tabla con los valores del experimento
tabla = pd.DataFrame({'Experimento: ': l})

print(tabla.to_string(index=False))
```

```python
#Grafica de Convergencia de Montecarlo Crudo
plt.plot(l,color="b")
plt.title("Convergencia de MC Crudo")
plt.xlabel("Tamaño de la Muestra")
plt.ylabel("Valor de la Media Muestral")
plt.grid(True)
plt.show()
```
